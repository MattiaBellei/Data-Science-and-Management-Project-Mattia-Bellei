```python
# creo una funzione di nome plot_column che crea il grafico a barre di come sono distribuiti i dati di quella colonna
import matplotlib.ticker as ticker



def plot_column(column_name):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column_name, data=combined_data)
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
 # utilizzo la funzione appena creata per una colonna di prova, ma è applicabile a tutte
plot_column('X1_ActualPosition')
```

```python
# creo una funzione che prenda come parametro una colonna e partendo da combined stampi i valori: media, mediana, varianza, deviazione standard e range

def analyze_column(column):
  column_array = column.to_numpy()

  mean_value = np.mean(column_array)
  median_value = np.median(column_array)
  variance_value = np.var(column_array)
  std_deviation = np.std(column_array)
  range = np.max(column_array) - np.min(column_array)

  print(f'Column: {column.name}\n')
  print(f'Mean: {mean_value}')
  print(f'Median: {median_value}')
  print(f'Variance: {variance_value}')
  print(f'Standard Deviation: {std_deviation}')
  print(f'Range: {range}')

# Esempio di prova della funzione
analyze_column(combined_data['X1_ActualPosition'])
```

```python
# creo una funzione che prenda come parametro il nome di una colonna e che per quella colonna calcoli la media dei suoi valori in ogni singolo dataset
# di esperimento (experiment_nn_data). La funzione poi stampa un grafico con le medie di tutti i diversi esperimenti

def plot_experiment_averages(column_name):
    averages = []
    experiment_numbers = []
    for i in range(1, 19):
      if i < 10:
        experiment_num_str = f'experiment_0{i}'
      else:
        experiment_num_str = f'experiment_{i}'
      df= pd.read_csv(f'{root_path}/{experiment_num_str}.csv')
      avg = df[column_name].mean()
      averages.append(avg)
      experiment_numbers.append(i)


    plt.figure(figsize=(10, 6))
    plt.bar(experiment_numbers, averages)
    plt.xlabel('Experiment Number')
    plt.ylabel(f'Mean of {column_name}')
    plt.xticks(experiment_numbers)
    plt.title(f'Mean of {column_name} in all experiments')
    plt.grid(True, alpha=0.7)
    plt.show()
```

```python
#provo funzione
plot_experiment_averages('X1_ActualPosition')
```

```python
# creo una funzione che prenda come parametro il nome di una colonna e che per quella colonna calcoli una statistica dei suoi valori in ogni dataset
# di esperimento (experiment_nn_data). Poi questa funzione deve stampare un grafico con le statistiche di tutti i diversi esperimenti

def plot_experiment(column_name, value):
    value_list = []
    experiment_numbers = []
    for i in range(1, 19):
      if i < 10:
        experiment_num_str = f'experiment_0{i}'
      else:
        experiment_num_str = f'experiment_{i}'
      df= pd.read_csv(f'{root_path}/{experiment_num_str}.csv')
      if value == 'Median':
        output = df[column_name].median()
      elif value == 'Max':
        output = df[column_name].max()
      elif value == 'Min':
        output = df[column_name].min()
      elif value == 'Var':
        output = df[column_name].var()
      elif value == 'Std':
        output = df[column_name].std()
      value_list.append(output)
      experiment_numbers.append(i)

    plt.figure(figsize=(10, 6))
    plt.bar(experiment_numbers, value_list)
    plt.xlabel('Experiment Number')
    plt.ylabel(f'{value} of {column_name}')
    plt.xticks(experiment_numbers)
    plt.title(f'{value} of {column_name} in all experiments')
    plt.grid(True, alpha=0.7)
    plt.show()
```

```python
# provo funzione appena creata
plot_experiment('X1_ActualPosition', 'Var')
```
